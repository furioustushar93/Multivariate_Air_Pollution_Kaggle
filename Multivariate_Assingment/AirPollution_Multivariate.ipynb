```python
dataset = pd.read_csv("Train.csv",encoding='utf-8')
```

```python
X = dataset.values[:,:5]
Y = dataset.target
print(X.shape)
Y = Y.values.reshape((-1,1))
print(Y.shape)
print(X.shape)

#Normalize
X= (X-X.mean())/X.std()
ones = np.ones((X.shape[0],1))
X = np.hstack((ones,X))
print(X.shape)
X[:,:5]
```

```python
def hypothesis(X,theta):
    hx = np.dot(X,theta)
    return hx
    
```

```python
def error(X,Y,theta):
    m = X.shape[0]
    error = 0
    
    for i in range(m):
        hx = hypothesis(X[i],theta)
        error += (hx-Y[i])**2 
    return error/(2*m)

def gradient(X,Y,theta):
    grad = np.zeros((X.shape[1]))
    m = X.shape[0]
    n = X.shape[1]
    for j in range(n):
        for i in range(m):
            hx = hypothesis(X[i],theta)
            grad[j] += ((Y[i] - hx)*X[i][j]) 
    return grad

def algo(X,Y,learning_rate=0.001):
    theta = np.ones((X.shape[1]))
    error_list = []
    itr = 0
    max_itr = X.shape[0]
    while(itr<max_itr):
        grad = gradient(X,Y,theta)
        e = error(X,Y,theta)
        error_list.append(e)
        theta = theta + learning_rate*grad
        itr += 1
    return theta,error_list
```

```python
final_theta,error = algo(X,Y)
```

# Batch Gradient Implementation 

```python
def error_batch(X,Y,theta):
    m = X.shape[0]
    total_e = 0
    
    for i in range(m):
        h = hypothesis(X[i],theta)
        e = (Y[i] - h)**2
        total_e += e
        
    return total_e/(2*m)

def Batch_gradient(X,Y,theta,batch_size=10):
    grad = np.zeros((X.shape[1]))
    m = X.shape[0]
    n = X.shape[1]
    indices = np.arange(m)
    np.random.shuffle(indices)
    indices = indices[:batch_size]
    for j in range(n):
        for i in indices:
            hx = hypothesis(X[i],theta)
            grad[j] += ((Y[i] - hx)*X[i][j])
    return grad

def Batch_GradientDescent(X,Y,learning_rate = 0.001):
    theta = np.ones((X.shape[1]))
    itr = 0
    e = error_batch(X,Y,theta)
    error_list = [e]
    while(itr<X.shape[0]):
        grad = Batch_gradient(X,Y,theta)
        theta = theta + learning_rate*grad
        ne = error_batch(X,Y,theta)
        
        if(itr%10):
            error_list.append(ne)
            
        de = ne - e
        e = ne
        itr += 1
    return theta,error_list
```
